# 2 Plotting a Function
Goal: Write a **function** that will plot $f(x) = x^2$ and $g(x) = \sqrt{x}$ on the same plot and return the $f(x)$ and $g(x)$ values as separate lists. Your plot needs to have the following:
- An appropriate figure size (`figsize=(10,8)` is recommended)
- Labeled x and y axis
- A title
- Different colors and linestyles for each function
- A legend
- Your domain should be $x \in [0, 10]$. *Hint: Use `np.linspace` to set a range of values for your functions.*

Your function should return:
- A list of $f(x)$ values
- A list of $g(x)$ values.

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_functions():
    # Define the domain of x from 0 to 10
    x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
    
    # Define the functions f(x) = x^2 and g(x) = sqrt(x)
    f_x = x**2
    g_x = np.sqrt(x)
    
    # Create a plot with figsize (10, 8)
    plt.figure(figsize=(10, 8))
    
    # Plot f(x) = x^2 with blue color and solid line
    plt.plot(x, f_x, color='blue', linestyle='-', label=r'$f(x) = x^2$')
    
    # Plot g(x) = sqrt(x) with green color and dashed line
    plt.plot(x, g_x, color='green', linestyle='--', label=r'$g(x) = \sqrt{x}$')
    
    # Add title to the plot
    plt.title('Plot of f(x) = x^2 and g(x) = sqrt(x)', fontsize=16)
    
    # Label the x-axis and y-axis
    plt.xlabel('x', fontsize=14)
    plt.ylabel('y', fontsize=14)
    
    # Add a legend to the plot
    plt.legend()
    
    # Show gridlines for better readability
    plt.grid(True)
    
    # Display the plot
    plt.show()
    
    # Return the values of f(x) and g(x) as separate lists
    return f_x.tolist(), g_x.tolist()

# Call the function and get the f(x) and g(x) values
f_values, g_values = plot_functions()
```

# 3 Introduction to Subplots
Goal: Write a **module** (name it whatever you want, as long as it is professional) that contains two functions that do the following: 

Function 1: 
- Write a **function** that will plot two subplots side-by-side (horizontal).
- The **left** subplot will be: $h(x) = \cos(x)$.
- The **right** subplot will be $k(x) = \sin(x)$.
- Include a title, axis labels and a domain of $x \in [0, 2 \pi]$ (this should NOT be hardcoded)

Function 2:
- Exactly the same as your first function, **EXCEPT** the subplots will be on top of each other (vertical). 

```python
import numpy as np
import matplotlib.pyplot as plt

# Function 1: Side-by-side subplots (horizontal layout)
def plot_horizontal_subplots(domain):
    """
    This function creates two subplots side by side:
    1. The left subplot: h(x) = cos(x)
    2. The right subplot: k(x) = sin(x)
    
    Parameters:
    - domain: A tuple or list representing the start and end values for the domain of x (e.g., (0, 10))
    """
    
    # Generate x values based on the domain
    x = np.linspace(domain[0], domain[1], 15)  # 15 points between domain[0] and domain[1]
    
    # Define the functions
    h_x = np.cos(x)  # h(x) = cos(x)
    k_x = np.sin(x)  # k(x) = sin(x)
    
    # Create the subplots (side-by-side)
    fig, axs = plt.subplots(1, 2, figsize=(14, 6))  # 1 row, 2 columns
    
    # Left subplot: h(x) = cos(x)
    axs[0].plot(x, h_x, color='blue', label=r'$h(x) = \cos(x)$')
    axs[0].set_title(r'$h(x) = \cos(x)$', fontsize=14)
    axs[0].set_xlabel('x', fontsize=12)
    axs[0].set_ylabel('h(x)', fontsize=12)
    axs[0].legend()
    axs[0].grid(True)
    
    # Right subplot: k(x) = sin(x)
    axs[1].plot(x, k_x, color='green', label=r'$k(x) = \sin(x)$')
    axs[1].set_title(r'$k(x) = \sin(x)$', fontsize=14)
    axs[1].set_xlabel('x', fontsize=12)
    axs[1].set_ylabel('k(x)', fontsize=12)
    axs[1].legend()
    axs[1].grid(True)
    
    # Set a common x-axis limit for both subplots
    axs[0].set_xlim(domain[0], domain[1])
    axs[1].set_xlim(domain[0], domain[1])
    
    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()


# Function 2: Stacked subplots (vertical layout)
def plot_vertical_subplots(domain):
    """
    This function creates two subplots stacked on top of each other:
    1. The top subplot: h(x) = cos(x)
    2. The bottom subplot: k(x) = sin(x)
    
    Parameters:
    - domain: A tuple or list representing the start and end values for the domain of x (e.g., (0, 10))
    """
    
    # Generate x values based on the domain
    x = np.linspace(domain[0], domain[1], 15)  # 15 points between domain[0] and domain[1]
    
    # Define the functions
    h_x = np.cos(x)  # h(x) = cos(x)
    k_x = np.sin(x)  # k(x) = sin(x)
    
    # Create the subplots (stacked vertically)
    fig, axs = plt.subplots(2, 1, figsize=(8, 12))  # 2 rows, 1 column
    
    # Top subplot: h(x) = cos(x)
    axs[0].plot(x, h_x, color='blue', label=r'$h(x) = \cos(x)$')
    axs[0].set_title(r'$h(x) = \cos(x)$', fontsize=14)
    axs[0].set_xlabel('x', fontsize=12)
    axs[0].set_ylabel('h(x)', fontsize=12)
    axs[0].legend()
    axs[0].grid(True)
    
    # Bottom subplot: k(x) = sin(x)
    axs[1].plot(x, k_x, color='green', label=r'$k(x) = \sin(x)$')
    axs[1].set_title(r'$k(x) = \sin(x)$', fontsize=14)
    axs[1].set_xlabel('x', fontsize=12)
    axs[1].set_ylabel('k(x)', fontsize=12)
    axs[1].legend()
    axs[1].grid(True)
    
    # Set a common x-axis limit for both subplots
    axs[0].set_xlim(domain[0], domain[1])
    axs[1].set_xlim(domain[0], domain[1])
    
    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()
```
